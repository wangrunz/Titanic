```python
import pandas as pd
import numpy as np
df = pd.read_csv('cheat.csv')
dt = pd.read_csv('test.csv')
```

```python
def form(string):
    ll=string.split()
    s=ll[0]+'. '
    for i in range(1,len(ll)):
        s+=ll[i]+' '
    s=s[:-1]
    return s
```

```python
df.loc[df['survived']=='†', 'survived'] = 0
df['survived']=df['survived'].fillna(1)
df['Name']=df['Surname']+', '+df['First Names'].apply(form)
df[['Name','survived']]
```
